```python
import streamlit as st
from streamlit.server.server import Server
from web3 import Web3

# Connect to Ethereum network using Infura
web3 = Web3(Web3.HTTPProvider('https://mainnet.infura.io/v3/YOUR_INFURA_PROJECT_ID'))

# Define the contract ABI and address
contract_address = '0x...your_contract_address...'
contract_abi = [
    #add contract abi
]

# Create the contract instance
contract = web3.eth.contract(address=contract_address, abi=contract_abi)

# Define App 1
def app1():
    st.title('Inheritance Smart Contract')

    company = st.text_input('Company')
    ticker = st.text_input('Ticker')
    shares = st.number_input('Shares')
    brokerage = st.text_input('Brokerage')
    tokenURI = st.text_input('Token URI')

    if st.button('Register Stock'):
        tokenId = contract.functions.registerStock(company, ticker, shares, brokerage, tokenURI).transact()
        st.success('Token ID: {}'.format(tokenId))

# Define App 2
def app2():
    st.title('My Beneficiary')

    name = st.text_input("Enter beneficiary's name:")
    age = st.number_input("Enter beneficiary's age:", min_value=0, max_value=150)
    ssn = st.text_input("Enter beneficiary's SSN:")

    if name and age and ssn:
        # Perform My Beneficiary
        my_beneficiary = contract.functions.my_beneficiary(name, age, ssn).call()

        # Display the Beneficiary details
        st.subheader("Beneficiary Details")
        st.write(my_beneficiary)

def main():
    # Create a form for login information
    st.header("User Login")
    username = st.text_input("Username")
    password = st.text_input("Password", type="password")
    remember_me = st.checkbox("Remember Me")

    # Store the login information
    if st.button("Login"):
        if username and password:
            # Store the login credentials in a database or file
            # You can implement your own logic here to save the credentials
            # For simplicity, we'll just display the credentials for now
            st.write("Username:", username)
            st.write("Password:", password)
            st.write("Remember Me:", remember_me)
        else:
            st.warning("Please enter both username and password")

    # Launch App 1
    server1 = Server()
    server1.start(app1)

    # Launch App 2
    server2 = Server()
    server2.start(app2)

    # Connect App 1 to App 2
    app2_url = server2.get_url('/')
    server1.set_environment_info("app2_url", app2_url)

    # Start the Streamlit server
    Server.start(primary_module=None, is_manually_stopped=False)

if __name__ == "__main__":
    main()
```
